```python
import os
import shutil
from PIL import Image

results_dir = "<results of CycleGan>"
output_dir = "<Path to output directory-Domain B enhanced>"
original_dir = "<Path to original Domain B directory>"

os.makedirs(output_dir, exist_ok=True)


orig_ext = {}
for fname in os.listdir(original_dir):
    base, ext = os.path.splitext(fname)
    orig_ext[base] = ext  

for filename in os.listdir(results_dir):
    if filename.endswith("_fake_B.png"):
        base = filename.replace("_fake_B.png", "")

        if base not in orig_ext:
            print(f"âš  Warning: {base} not found in original dataset!")
            continue
        
        target_ext = orig_ext[base]  
        target_name = base + target_ext
        dst_path = os.path.join(output_dir, target_name)
        
        src_path = os.path.join(results_dir, filename)
        
        # Convert PNG â†’ JPG
        img = Image.open(src_path).convert("RGB")
        img.save(dst_path)
        
        print(f"Saved: {dst_path}")

print("\nâœ… Done! Images converted to original .jpg filenames.")
```
